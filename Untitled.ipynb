```python
x= []
y=[]
for i in range(270):
    x.append(16*np.random.random())
    y.append(9*np.random.random())
```

```python
df = pd.DataFrame({"X": x, "Y": y})
df["logX"] = df["X"].apply(np.log)
df["logY"] = df["Y"].apply(np.log)
```

```python
df.plot.scatter(x="X", y="logY")
```

```python
df.plot.hexbin(x="X", y="logY")
```
